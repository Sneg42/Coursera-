```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree

data = pd.read_csv('titanic.csv', index_col='PassengerId')
data[:10]
```

```python
dataForLearning = data[['Pclass', 'Fare', 'Age', 'Sex', 'Survived']].dropna()
target = dataForLearning['Survived'].values

dataForLearning= dataForLearning[['Pclass', 'Fare', 'Age', 'Sex']]
dataForLearning = dataForLearning.replace(to_replace=['male', 'female'], value=[1, 0])
dataForLearning = dataForLearning.values


test_index = np.arange(0,4,1)

train_dataForLearning = np.delete(dataForLearning, test_index, axis=0)
train_target = np.delete(target, test_index)


testing_dataForLearning = dataForLearning[test_index]
testing_target = target[test_index]


clf = tree.DecisionTreeClassifier()
clf.fit(train_dataForLearning, train_target)
importances = clf.feature_importances_
print (importances)
print (testing_target)
print (clf.predict(testing_dataForLearning))
```
