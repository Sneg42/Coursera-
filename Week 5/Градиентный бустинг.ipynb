```python
import numpy as np
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import log_loss
```

```python
df = pd.read_csv('gbm-data.csv')
df[:2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Activity'], axis=1).values, df['Activity'].values,
                                                    test_size=0.8, random_state=241)
```

```python
learning_rate = [1, 0.5, 0.3, 0.2, 0.1]
for i in learning_rate:
    gb = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=i)
    gb.fit(X_train, y_train)
    print (gb.staged_decision_function(X_train))
#     gb.staged_decision_function(X_test)
    
#     gb.predict_proba(X_train)
```
