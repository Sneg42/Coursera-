```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_recall_curve
import itertools
```

```python
classification = pd.read_csv('classification.csv')
classification[:2]
```

```python
TP = 0
FP = 0
FN = 0
TN = 0
for index, i in enumerate(classification['pred']):
    if i > classification['true'].iloc[index]:
        FP +=1
    if i < classification['true'].iloc[index]:
        
        FN +=1
    if i == 1 and i == classification['true'].iloc[index]:
        
        TP +=1
    if i == 0 and i == classification['true'].iloc[index]:
        TN +=1    
print ('TP - {} FP - {}\nFN - {} TN - {}'.format(TP, FP, FN, TN))   
print (TP + FP + FN + TN)
```

```python
accuarcy = accuracy_score(classification['true'], classification['pred'])
precision = precision_score(classification['true'], classification['pred'])
recall = recall_score(classification['true'], classification['pred'], average='binary')
f1 = f1_score(classification['true'], classification['pred'])
print (accuarcy)
print (precision)
print (recall)
print (f1)
```
