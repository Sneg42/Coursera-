```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn import svm, preprocessing
from sklearn.neighbors import KNeighborsClassifier 

from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression 

data = pd.read_csv('wine.csv', names=['class', 'a', 'b', 'c', 'd', 't', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm'])

mas =[]

#генератор разбиения перекрестной проверки (на 5 блоков и перемещать)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

X = data.drop(['class'],  axis=1)
#Масштабирование 
X = preprocessing.scale(X)
y = data['class']

for i in range(1, 51):
    
    clf = KNeighborsClassifier(n_neighbors=i)
    scores = cross_val_score(clf, X, y, cv=kfold)
    mas.append(np.mean(scores))
    
print (np.mean(mas), mas.index(max(mas)))
```

```python
from sklearn.datasets import load_boston

bos = load_boston()


X = bos.data.astype(int)

X = preprocessing.scale(X)
y = bos.target.astype(int)



kfold = KFold(n_splits=5, shuffle=True, random_state=42)
maximum = []
number_p = []
for i in np.linspace(1, 10, 200):
    clf = KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=i , weights='distance')
    score = np.mean(cross_val_score(clf, X, y, cv=kfold, scoring='neg_mean_absolute_error'))
    
    maximum.append(np.mean(score))
    
    number_p.append(i)

print (number_p[maximum.index(min(maximum))])
print (len(number_p))
    
```
